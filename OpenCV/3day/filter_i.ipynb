```python
mx=np.max(filtered)
mi=np.min(filtered)

n=(filtered-mi)/(mx-mi)*255 #-2 ~ 0  => 0~255
print(n)

n2=np.abs(filtered)/np.max(np.abs(filtered))*255   #0~2 => 0~255
print(n2)

norm=cv2.normalize(filtered,None ,0,255,cv2.NORM_MINMAX).astype("uint8")
print(norm)
```

```python
src = cv2.imread("lena.png", cv2.IMREAD_GRAYSCALE).astype("float")

kernel = np.array([[0,1,0],[1,0,1],[0,1,0]])

filtered=cv2.filter2D(src,-1,kernel)
print(np.max(filtered))
print(np.min(filtered))

cv2.imwrite("output1.jpg",filtered.astype("uint8"))

```

```python
src = cv2.imread("lena.png", cv2.IMREAD_GRAYSCALE).astype("float")

kernel = np.array([[0,1,0],[0,0,0],[0,-1,0]])

filtered=cv2.filter2D(src,-1,kernel)+128
print(np.max(filtered))
print(np.min(filtered))

cv2.imwrite("output3.jpg",filtered.astype("uint8"))

```

```python
src2 = cv2.imread("lena.png", cv2.IMREAD_GRAYSCALE).astype("int16")

kernel = np.array([[0,1,0],[0,0,0],[0,-1,0]])

filtered=cv2.filter2D(src,-1,kernel)
filtered2=cv2.normalize(filtered,None,0,255,cv2.NORM_MINMAX)
print(np.max(filtered2))
print(np.min(filtered2))

cv2.imwrite("output2.jpg",filtered.astype("uint8"))

```

```python
src2 = cv2.imread("face.png", cv2.IMREAD_GRAYSCALE).astype("int16")

kernel = np.array([[0,1,0],[0,0,0],[0,-1,0]])

filtered=cv2.filter2D(src2,-1,kernel)
filtered2=cv2.normalize(filtered,None,0,255,cv2.NORM_MINMAX)
print(np.max(filtered2))
print(np.min(filtered2))

cv2.imwrite("output4.jpg",filtered.astype("uint8"))
```
