```python
#n 진수 게임

def solution(n, t, m, p):
    numbers = []
    answer =[]
           
    for i in range(0,t*m):
        a = numeral_system(i,n)
        numbers.append(a)
    numbers_a=  list(''.join(numbers))      
    
    for j in range(p-1,t*m,m):
        answer.append(numbers_a[j])
        
    answer=''.join(answer)
    return answer

rotation = '0123456789ABCDEF'

def numeral_system(number , base):
    q, r = divmod(number, base)
    n = rotation[r]
    return numeral_system(q, base) + n if q else n
```
